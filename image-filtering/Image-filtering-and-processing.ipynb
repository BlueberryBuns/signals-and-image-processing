```python
import cv2 as cv
import numpy as np
import pandas as pd

from PIL import Image
from matplotlib import pyplot as plt
```

```python
image_name = input("Enter image name: ")
loaded_image = cv.imread(image_name)
if image_name.split('.')[1] == 'jpg':
    loaded_image = cv.cvtColor(loaded_image, cv.COLOR_BGR2RGB)

fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(15,15))

ax.imshow(loaded_image);
ax.set_title(image_name.split(".")[0])
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

```python
a = [int(a) for a in input("Enter coordinates of the first point (eg. 100,100): ").split(',')]
b = [int(b) for b in input("Enter coordinates of the second point (eg. 200,200): ").split(',')]
```

```python
cropped_img = loaded_image[a[1]:b[1],a[0]:b[0]]
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(15,15))
ax.imshow(cropped_img)
ax.set_title(image_name.split(".")[0])
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()

new_file_name = input('Enter new file name: ')
if new_file_name.split('.')[1] == 'jpg':
    cropped_img = cv.cvtColor(cropped_img, cv.COLOR_BGR2RGB)
if cv.imwrite(new_file_name, cropped_img):
    print("File saved successfully")
else: print("Error during loading the file")
```
